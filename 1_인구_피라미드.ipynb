# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
file_name = '201511_연령별인구현황.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col=1, usecols='B,E:O')
df_m.head(3)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int) # 1,521,194->1521194
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관',usecols='B,R:AB')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns #컬럼명 통일
```

```python
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int) # 1,521,194->1521194
df_w.iloc[0]
```

## 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000) #단위 : 천명
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title('2015년 11월 대한민국 인구 피라미드')
plt.savefig('2015년 11월 인구피라미드.png',dpi=100)
plt.show()
```
