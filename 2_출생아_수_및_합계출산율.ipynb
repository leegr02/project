# 2. 출생아 수 및 합계출산율
- 합계출산율이란? 한 여자가 가임기간(15~49세)에 낳을 것으로 기대되는 평균 출생아 수
- 구글검색: 출상아수 합계출산율
- e-나라지표

```python
import pandas as pd
file_name = '출생아수및합계출산율.xlsx'
df = pd.read_excel(file_name, skiprows=2, nrows=2, index_col=0)
df
```

```python
df.rename(index={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'}, inplace=True)
```

```python
df = df.T #행과 열을 바꿈
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
#matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
ax1.plot(df.index, df['출생아 수'], color='#ff812d')
ax2 = ax1.twinx() #x축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(200,520)
ax1.set_yticks([200,300,400])
ax1.bar(df.index, df['출생아 수'], color='skyblue')

ax2 = ax1.twinx() #x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율(가입여성 1명당 명')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,0.4,0.8,1.2])
ax2.plot(df.index, df['합계 출산율'], color='orange')
```

```python
fig, ax1 = plt.subplots(figsize=(13,5))
fig.suptitle('출상아 수 및 합계 출산율')
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(200,520)
ax1.set_yticks([200,300,400])
ax1.bar(df.index, df['출생아 수'], color='skyblue')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val+5,val,ha='center')

ax2 = ax1.twinx() #x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율(가입여성 1명당 명')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,0.4,0.8,1.2])
ax2.plot(df.index, df['합계 출산율'], color='orange', marker='o', ms=15, lw=5, mec='white', mew=3)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val+0.08,val,ha='center')
```
